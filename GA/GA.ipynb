```python
class Solving:
  def __init__(self,bit_string):
    self.bit_string = bit_string
    self.fitness = self.calculate_fitness()

  def check_genotype(self):
    return int(self.bit_string,2)

  def check_phenotype(self):
    return self.check_genotype()-11

  def calculate_fitness(self):
    x = self.check_phenotype()
    return 9 - x - (7*x*x) - (2*x*x*x)

  def recalculate_fitness(self):
    self.fitness = self.calculate_fitness()

solutions = ["011011","100010","000100","111001"]

sols = []

for sol in solutions:
  sols.append(Solving(sol))
  print("Bits", sol)
  print("Genotype", sols[-1].check_genotype())
  print("Phenotype", sols[-1].check_phenotype())
  print("Fitness func", sols[-1].calculate_fitness())
```

```python
import random


class Environment:
    def __init__(self, sols, n_generation, mutate_rate):
        self.sols = sols
        self.n_generation = n_generation
        self.mutate_rate = mutate_rate

    def add_new_solvs(self):
        first_parent, second_parent = random.sample(self.sols, 2)
        splitter = random.randint(1, len(first_parent.bit_string) - 1)
        new_solv = Solving(first_parent.bit_string[:splitter] + second_parent.bit_string[splitter:])
        self.sols.append(new_solv)

    def crossingover(self):
        self.add_new_solvs()
        self.add_new_solvs()

        self.mutation()

        self.sols = sorted(self.sols, key=lambda x: x.fitness, reverse=True)[:4]

        for sol in self.sols:
            print(sol.bit_string)
            print(sol.fitness)

    def mutation(self):
        for sol in self.sols:
            # print("Was", sol.bit_string)
            bit_list = list(sol.bit_string)
            for i in range(len(bit_list)):
                rand = random.random()
                if rand < self.mutate_rate:
                    bit_list[i] = "0" if bit_list[i] == "1" else "1"
            sol.bit_string = ''.join(bit_list)
            # print("Became", sol.bit_string)
            sol.recalculate_fitness()

    def run(self):
      for _ in range(self.n_generation):
        self.crossingover()


      best = sorted(self.sols, key=lambda x: x.fitness, reverse=True)[0]
      return best


env = Environment(sols, 100,1/25)
best = env.run()
print("Fitness", best.fitness)
print("Bit mask", best.bit_string)
```
